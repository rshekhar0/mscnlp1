```python
import nltk  # Importing the NLTK library
from nltk.tag import DefaultTagger  # Importing DefaultTagger from nltk.tag module

# Downloading the 'treebank' resource
nltk.download('treebank')

# Creating a DefaultTagger with the default tag 'NN'
exptagger = DefaultTagger('NN')

# Evaluating the performance of the tagger using test sentences from the treebank corpus
from nltk.corpus import treebank
testsentences = treebank.tagged_sents()[1000:]  # Selecting test sentences from the treebank corpus
evaluation_result = exptagger.evaluate(testsentences)  # Evaluating the tagger
print("Tagger Evaluation Result:", evaluation_result)  # Printing the evaluation result

# Tagging a list of sentences using the DefaultTagger
import nltk  # Re-importing nltk module
from nltk.tag import DefaultTagger  # Re-importing DefaultTagger from nltk.tag module
exptagger = DefaultTagger('NN')  # Creating a new DefaultTagger instance
tagged_sentences = exptagger.tag_sents([['Hi', ','], ['How', 'are', 'you', '?']])  # Tagging the list of sentences
print("Tagged Sentences:", tagged_sentences)  # Printing the tagged sentences
```
