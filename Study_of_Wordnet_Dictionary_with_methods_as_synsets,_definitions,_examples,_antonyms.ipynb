```python
import nltk

# Download the WordNet resource silently
nltk.download('wordnet', quiet=True)

# Import the WordNet module from nltk.corpus
from nltk.corpus import wordnet

# Print the synsets (sets of synonymous words) associated with the word "computer"
print(wordnet.synsets("computer"))

# Print the definition of the synset representing the word "computer" (first noun sense)
print(wordnet.synset("computer.n.01").definition())

# Print examples of the word "computer" in context
print("Examples:", wordnet.synset("computer.n.01").examples())

# Get the antonyms (words with opposite meaning) of the word "buy" in its first verb sense
print(wordnet.lemma('buy.v.01.buy').antonyms())
```
