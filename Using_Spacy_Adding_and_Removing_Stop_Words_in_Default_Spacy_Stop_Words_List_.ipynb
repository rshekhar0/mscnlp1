```python
import spacy
import nltk
from nltk.tokenize import word_tokenize

# Load the English language model in Spacy
sp = spacy.load('en_core_web_sm')

# Get the default set of NLTK stop words
all_stopwords = sp.Defaults.stop_words

# Add the word "play" to the NLTK stop word collection
all_stopwords.add("play")

# Define the input text
text = "Rahul likes to play football, however he is not too fond of tennis."

# Tokenize the text using NLTK
text_tokens = word_tokenize(text)

# Remove stopwords from the tokens using the updated set of stop words
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords]
print(tokens_without_sw)

# Check if the word "not" exists in the stop word collection before trying to remove it
if 'not' in all_stopwords:
    all_stopwords.remove('not')

# Tokenize the text again
text_tokens = word_tokenize(text)

# Remove stopwords from the tokens using the updated set of stop words
tokens_without_sw = [word for word in text_tokens if not word in all_stopwords]
print(tokens_without_sw)
```
