```python
import nltk  # Importing the NLTK library
from nltk import tokenize  # Importing the tokenize module from NLTK
import os  # Importing the os module for redirecting output

# Redirecting NLTK download output to /dev/null
os.system("python -m nltk.downloader -q punkt")  # Quietly download the 'punkt' tokenizer models
os.system("python -m nltk.downloader -q words")  # Quietly download the 'words' dataset

# Input paragraph
para = "Hello! My name is Rahul Yadav. Today you'll be learning NLTK."

# Tokenizing the input paragraph into sentences
sents = tokenize.sent_tokenize(para)
print("\nsentence tokenization\n===================\n", sents)  # Printing the sentence tokenization results

# Word tokenization
print("\nword tokenization\n===================\n")
for index in range(len(sents)):  # Looping through each sentence
    words = tokenize.word_tokenize(sents[index])  # Tokenizing each sentence into words
    print(words)  # Printing the word tokenization results
```
