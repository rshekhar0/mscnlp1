```python
# Creating a dictionary by mapping words with their properties
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

# Printing the dictionary
print("Dictionary:", thisdict)

# Accessing and printing the value associated with the key "brand"
print("Brand:", thisdict["brand"])

# Getting and printing the length of the dictionary (number of key-value pairs)
print("Length of Dictionary:", len(thisdict))

# Printing the type of the dictionary
print("Type of Dictionary:", type(thisdict))
```
