```python
from gtts import gTTS  # Importing the gTTS class from the gtts module.
import os  # Importing the os module for interacting with the operating system.
import pygame  # Importing the pygame library for audio playback.

# Disable audio drivers
os.environ['SDL_AUDIODRIVER'] = 'dummy'  # Setting the SDL_AUDIODRIVER environment variable to 'dummy' to disable audio drivers.

mytext = "Hi This is Rahul and Welcome to Natural Language Processing"  # Defining the text to be converted to speech.
language = "en"  # Specifying the language for text-to-speech conversion.
myobj = gTTS(text=mytext, lang=language, slow=False)  # Creating a gTTS object for the given text and language.
myobj.save("myfile.mp3")  # Saving the generated audio file as "myfile.mp3".

# Initialize pygame mixer
pygame.mixer.init()  # Initializing the pygame mixer for audio playback.

# Check if the file exists before playing it
if os.path.exists("myfile.mp3"):  # Checking if the audio file exists.
    pygame.mixer.music.load("myfile.mp3")  # Loading the audio file for playback.
    pygame.mixer.music.play()  # Playing the loaded audio file.
    while pygame.mixer.music.get_busy():  # Looping until the audio playback finishes.
        pygame.time.Clock().tick(10)  # Limiting the loop to 10 ticks per second.
else:
    print("Error: Audio file not found.")  # Printing an error message if the audio file is not found.
```
