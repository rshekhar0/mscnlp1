```python
import nltk

# Import the WordNet module from nltk.corpus
from nltk.corpus import wordnet

# Print the synsets (sets of synonymous words) associated with the word "computer"
print(wordnet.synsets("computer"))

# Print the lemma names of the first noun sense of the word "computer"
print(wordnet.synset("computer.n.01").lemma_names())

# Iterate over all synsets of the word "computer" and print their lemma names
for synset in wordnet.synsets("computer"):
    print(f'{synset} --> {synset.lemma_names()}')

# Print all lemmas for the synset representing the word "computer" (first noun sense)
print(wordnet.synset('computer.n.01').lemmas())

# Get the synset corresponding to the lemma "computing_device"
print(wordnet.lemma('computer.n.01.computing_device').synset())

# Get the name of the lemma "computing_device"
print(wordnet.lemma('computer.n.01.computing_device').name())

# Get the hyponyms (more specific concepts) of the word "computer"
syn = wordnet.synset('computer.n.01')
print(syn.hyponyms)
print([lemma.name() for synset in syn.hyponyms() for lemma in synset.lemmas()])

# Calculate the lowest common hypernyms between "vehicle" and "car"
vehicle = wordnet.synset('vehicle.n.01')
car = wordnet.synset('car.n.01')
print(car.lowest_common_hypernyms(vehicle))
```
