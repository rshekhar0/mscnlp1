```python
from nltk.corpus import wordnet

# Print the synsets (sets of synonymous words) associated with the word "active"
print(wordnet.synsets("active"))

# Print the antonyms (words with opposite meaning) of the word "active" in its first adjective sense
print(wordnet.lemma('active.a.01.active').antonyms())
```
